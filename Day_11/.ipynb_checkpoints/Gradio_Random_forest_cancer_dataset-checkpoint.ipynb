```python
import gradio as gr
```

```python
from joblib import load
import numpy as np
```

```python
def predict(Clump,Unicell_size,Unicell_shape,Margadh,Septh,Bare_N,BChromantin,NeomM,Mitosis):
    tst=np.array([[Clump,Unicell_size,Unicell_shape,Margadh,Septh,Bare_N,BChromantin,NeomM,Mitosis]])
    bm_loaded=load("rf.joblib")
    if bm_loaded.predict(tst)[0]==0:
        return "Benign"
    else:
        return "Malignant"
```

```python
demo=gr.Interface(fn=predict,inputs=["number"]*9,outputs=["text"])
```

```python
demo.launch(share=True)
```
