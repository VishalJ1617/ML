```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ply
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,log_loss
from sklearn.model_selection import train_test_split,StratifiedKFold,RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,StackingClassifier,GradientBoostingClassifier
from sklearn.d
```

```python
img_df=pd.read_csv('Image_Segmention.csv')
```

```python
lbl=LabelEncoder()
```

```python
img_df['Class']=lbl.fit_transform(img_df['Class'])
```

```python
img_df
```

```python
img_df.info()
```

```python
X=img_df.drop('Class',axis=1)
```

```python
y=img_df['Class']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=23)
```

```python
svm=SVC(kernel='linear',random_state=23,probability=True)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
LDA=LinearDiscriminantAnalysis()
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
lr=LogisticRegression(random_state=23)
```

```python
from xgboost import XGBClassifier
```

```python
xgb=XGBClassifier(random_state=23)
```

```python
stack=StackingClassifier([('SVM',svm),('LDA',LDA),('GNB',gnb),('LR',lr)],final_estimator=xgb)
```

```python
stack.fit(X_train,y_train)
```

```python
stack.get_params()
```

```python
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

```python
params={'SVM__C':np.linspace(0.001,3,5),
        'LDA__solver':['svd', 'lsqr', 'eigen'],
        'GNB__var_smoothing':np.linspace(0.0001,0.01,5),
        'LR__l1_ratio':np.linspace(0.001,0.999,3),
       'LR__penalty': ['l1','l2','elasticnet',None],
       'LR__solver':['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga']}
```

```python
rgcv_stack=RandomizedSearchCV(stack,param_distributions=params,verbose=3,cv=kfold,scoring='neg_log_loss',n_iter=50)
```

```python
rgcv_stack.fit(X,y)
```

```python
rgcv_stack.best_params_
```

```python
rgcv_stack.best_score_
```
