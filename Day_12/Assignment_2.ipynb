##### 1. SVM without scaling and Linear kernel

```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
vehicle_df=pd.read_csv('Vehicle.csv')
```

```python
lbl=LabelEncoder()
```

```python
vehicle_df['Class']=lbl.fit_transform(vehicle_df['Class'])
```

```python
vehicle_df
```

```python
vehicle_df.info()
```

```python
X=vehicle_df.drop('Class',axis=1)
y=vehicle_df['Class']
```

```python
y
```

```python
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
```

```python
svm=SVC(kernel='linear',probability=True,random_state=23)
```

```python
svm.get_params()
```

```python
params={'C':np.linspace(0.001,5,5),'kernel':['linear']}
```

```python
gcv=GridSearchCV(svm,param_grid=params,cv=kfold,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
gcv.best_score_
```

#### 2. SVM without scaling and RBF kernel

```python
svm=SVC(kernel='rbf',probability=True,random_state=23)
```

```python
params={'C':np.linspace(0.001,5,30),'gamma':list(np.linspace(0.001,5,10))+['scale','auto']}
```

```python
gcv=GridSearchCV(svm,param_grid=params,cv=kfold,verbose=3,scoring='neg_log_loss')
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

##### 3. SVM with scaling and Linear kernel

```python
scaler=StandardScaler()
```

```python
pipe=Pipeline([('SCL',scaler),('SVM',svm)])
```

```python
pipe.get_params()
```

```python
params={'SVM__C':np.linspace(0.001,5,10),'SCL':[StandardScaler(),MinMaxScaler()]}
```

```python
# params={'SCL': [StandardScaler(),MinMaxScaler()],'SVM__kernel': ['linear'],'SVM__C':np.linspace(0.001,5,10),}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

## 4. SVM with scaling RBF

```python
scaler=StandardScaler()
```

```python
svm=SVC(kernel='rbf',probability=True,random_state=23)
```

```python
pipe=Pipeline([('SCL',scaler),('SVM',svm)])
```

```python
params={'SCL':[StandardScaler(),MinMaxScaler()],'SVM__C':np.linspace(0.001,5,30),
        'SVM__gamma':list(np.linspace(0.001,5,10))+['scale','auto']}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='neg_log_loss',verbose=3)
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```
