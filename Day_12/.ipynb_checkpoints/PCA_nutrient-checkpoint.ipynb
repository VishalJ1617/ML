```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
nutrient=pd.read_csv('nutrient.csv',index_col=0)
```

```python
nutrient
```

```python
scaler=StandardScaler()
```

```python
scaled_data=scaler.fit_transform(nutrient)
```

```python
prcomp=PCA()
```

```python
scores=prcomp.fit_transform(scaled_data)
```

```python
v_covar=np.cov(scaled_data,rowvar=False)
```

```python
v_covar
```

```python
values,vectors=np.linalg.eig(v_covar)
```

```python
print("Eigen vales =",values)
```

```python
print("variance ofo PC scores = ",np.var(scores,axis=0,ddof=1))
```

```python
prcomp.explained_variance_
```

```python
total_var=np.sum(prcomp.explained_variance_)
```

```python
print("% age variation")
print((prcomp.explained_variance_/total_var)*100)
```
