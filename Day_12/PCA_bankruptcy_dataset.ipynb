```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
backruptcy_df=pd.read_csv('bankruptcy.csv')
```

```python
backruptcy_df
```

```python
backruptcy_df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_data=scaler.fit_transform(backruptcy_df)
```

```python
from sklearn.decomposition import PCA
prcomp=PCA()
```

```python
prcomp.fit_transform(scaled_data)
```

```python
prcomp.explained_variance_
```

```python
total_var=np.sum(prcomp.explained_variance_)
```

```python
print((prcomp.explained_variance_/total_var)*100)
```
