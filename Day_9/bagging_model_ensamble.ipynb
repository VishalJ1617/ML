```python
import pandas as pd
import numpy as np
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
img_df=pd.read_csv('Image_Segmention.csv')
```

```python
X=img_df.drop('Class',axis=1)
```

```python
y=img_df['Class']
```

```python
from sklearn.model_selection import train_test_split,StratifiedKFold,GridSearchCV
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.7,random_state=23,stratify=y)
```

```python
lr=LogisticRegression(multi_class='multinomial')
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_prob=lr.predict_proba(X_test)
```

```python
from sklearn.metrics import log_loss
```

```python
lr_ll=log_loss(y_test,y_pred_prob)
```

```python
print('Alone logistic = ',lr_ll)
```

```python
bgc=BaggingClassifier(lr,n_estimators=30,random_state=23)
```

```python
bgc.fit(X_train,y_train)
```

```python
y_pred_prob=lr.predict_proba(X_test)
```

```python
bag_ll=log_loss(y_test,y_pred_prob)
```

```python
print("Bagging = ",bag_ll)
```

```python
from sklearn.svm import SVC
```

```python
svm=SVC(kernel='linear',probability=True,random_state=23)
```

```python
svm.fit(X_train,y_train)
```

```python
y_pred_prob=svm.predict_proba(X_test)
```

```python
svm_ll=log_loss(y_test,y_pred_prob)
```

```python
svm_ll
```

```python
bgc=BaggingClassifier(svm,n_estimators=15,random_state=23)
```

```python
bgc.fit(X_train,y_train)
```

```python
y_pred_prob=lr.predict_proba(X_test)
```

```python
bag_ll=log_loss(y_test,y_pred_prob)
```

```python
bag_ll
```

# Grid search cv models

```python
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

```python
lr=LogisticRegression()
```

```python
svm=SVC(kernel='linear',probability=True,random_state=23)
```

```python
dtc=DecisionTreeClassifier(random_state=23)
```

```python
bagg=BaggingClassifier(random_state=23)
```

```python
bagg.get_params()
```

```python
params={ 'estimator':[lr,svm,dtc],'n_estimators':[10,15,30,40,50]}
```

```python
gcv=GridSearchCV(bagg,param_grid=params,cv=kfold,verbose=3,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```
