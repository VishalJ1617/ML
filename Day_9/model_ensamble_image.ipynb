```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, log_loss
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split,StratifiedKFold,GridSearchCV
```

```python
image=pd.read_csv('Image_Segmention.csv')
```

```python
lbl=LabelEncoder()
```

```python
image['Class']=lbl.fit_transform(image['Class'])
```

```python
X=image.drop('Class',axis=1)
y=image['Class']
```

```python
svm=SVC(probability=True,random_state=23)
```

```python
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

```python
svm.get_params()
```

```python
params={'gamma': ['scale','auto'],'C':np.linspace(0.001,5,5)}
```

```python
gcv=GridSearchCV(svm,param_grid=params,verbose=3,cv=kfold,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
svm_ll=gcv.best_score_
```

# logistic regression

```python
lr=LogisticRegression()
```

```python
lr.get_params()
```

```python
params={'penalty':['l1','l2','elasticnet',None],'multi_class':['multimodel','ovr']}
```

```python
gcv=GridSearchCV(lr,param_grid=params,verbose=3,cv=kfold,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
lr_ll=gcv.best_score_
```

# Gaussian nb

```python
nb=GaussianNB()
```

```python
nb.get_params()
```

```python
params={'var_smoothing':np.linspace(0.0001,0.999,5)}
```

```python
gcv=GridSearchCV(nb,param_grid=params,verbose=3,cv=kfold,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
nb_ll=gcv.best_score_
```

# esemble without weights

```python
voting=VotingClassifier(estimators=[('SVM',svm),('LR',lr),('NB',nb)],voting="soft")
```

```python
voting.get_params()
```

```python
params={'SVM__gamma': ['scale','auto'],'SVM__C':np.linspace(0.001,5,5),'LR__penalty':['l1','l2','elasticnet',None],'LR__multi_class':['multimodel','ovr'] ,'NB__var_smoothing':np.linspace(0.0001,0.999,5)}
```

```python
gcv=GridSearchCV(voting,param_grid=params,verbose=3,cv=kfold,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
unweighted_ll=gcv.best_score_
```

```python
print("unweighted",unweighted_ll)
print("SVM",svm_ll)
print("LR",lr_ll)
print("NB",nb_ll)
```

# weighted ensemble

```python
voting=VotingClassifier(estimators=[('SVM',svm),('LR',lr),('NB',nb)],voting="soft",weights=[0.5,0.4,0.1])
```

```python
params={'SVM__gamma': ['scale','auto'],'SVM__C':np.linspace(0.001,5,5),'LR__penalty':['l1','l2','elasticnet',None],'LR__multi_class':['multimodel','ovr'] ,'NB__var_smoothing':np.linspace(0.0001,0.999,5)}
```

```python
gcv=GridSearchCV(voting,param_grid=params,verbose=3,cv=kfold,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
gcv.best_score_
```
