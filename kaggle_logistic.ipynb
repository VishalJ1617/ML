```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold,GridSearchCV
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.metrics import log_loss,accuracy_score
```

```python
train=pd.read_csv('train.csv')
train
```

```python
X=train.drop('quality',axis=1)
y=train['quality']
```

```python
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

```python
lr=LogisticRegression()
```

```python
params={'penalty':['l1','l2','elasticnet',None],'solver':['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga'],'multi_class':['ovr','multinomial']}
```

```python
gcv=GridSearchCV(lr,param_grid=params,cv=kfold,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
test=pd.read_csv('test.csv')
```

```python
lr=LogisticRegression(penalty='l2',multi_class='multinomial',solver= 'newton-cg')
```

```python
lr.fit(X,y)
```

```python
y_pred=lr.predict(test)
solution=pd.DataFrame(y_pred)
```
