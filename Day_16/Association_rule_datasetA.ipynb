```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
fi=[]

with open('DataSetA.csv','r') as file:
    fi=file.read()
fi=fi.split('\n')
```

```python
dataset=[]
for i in fi:
    dataset.append(i.split(','))
```

```python
dataset
```

```python
te=TransactionEncoder()
te_ary=te.fit_transform(dataset)
te_ary
```

```python
fp_df=pd.DataFrame(te_ary,columns=te.columns_)
```

```python
fp_df=fp_df.drop('',axis=1)
```

```python
itemset=apriori(fp_df,min_support=0.2,use_colnames=True)
```

```python
itemset
```

```python
rule=association_rules(itemset,metric='confidence',min_threshold=0.45)
```

```python
rule
```

```python
rule=rule[['antecedents','consequents','support','lift']]
```

```python
rule=rule[rule['lift']>1]
```

```python
rule.sort_values(by='lift',ascending=False)
```
