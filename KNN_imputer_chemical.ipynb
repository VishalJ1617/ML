```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import ElasticNet
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold, GridSearchCV
```

```python
chem_df=pd.read_csv('ChemicalProcess.csv')
```

```python
X=chem_df.drop('Yield',axis=1)
y=chem_df['Yield']
```

##### Imputing with mean stratrgy

```python
imputer=KNNImputer( n_neighbors=1)
```

```python
elastic = ElasticNet()
```

##### Applying pipeline for imputation and elasticnet

```python
pipe=Pipeline([('IMPUTER',imputer),('ELASTIC',elastic)])
```

```python
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
```

```python
l1_rat=np.linspace(0.001,0.999,20)
```

```python
alphas=np.linspace(0.001,4,20)
```

```python
pipe.get_params()
```

```python
params={'ELASTIC__alpha':alphas,'ELASTIC__l1_ratio':l1_rat,'IMPUTER__n_neighbors':np.arange(1,10)}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold)
```

```python
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
```

```python
print(gcv.best_score_)
```
