```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import ElasticNet
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.ensemble import RandomForestRegressor,BaggingRegressor
```

```python
chem_df=pd.read_csv('ChemicalProcess.csv')
X=chem_df.drop('Yield',axis=1)
y=chem_df['Yield']
```

```python
imp=SimpleImputer()
```

```python
rf=RandomForestRegressor(random_state=23)
```

```python
pipe=Pipeline([('IMPUTER',imp),('RF',rf)])
```

```python
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
```

```python
pipe.get_params()
```

```python
params={'IMPUTER__strategy':['mean','median'],'RF__max_features':[3,4,5,6,8]}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='r2')
```

```python
gcv.fit(X,y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

# 2. Imputer bagging elastic

```python
imp=SimpleImputer()
```

```python
elastic=ElasticNet(random_state=23)
```

```python
bgr=BaggingRegressor(elastic,random_state=23)
```

```python
pipe=Pipeline([('IMPUTER',imp),('BGR',bgr)])
```

```python
pipe.get_params()
```

```python
params={'IMPUTER__strategy':['mean','median'],'BGR__estimator__l1_ratio':np.linspace(0.001,4,10),'BGR__estimator__alpha':np.linspace(0.001,0.999,10)}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='r2')
```

```python
gcv.fit(X,y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```
