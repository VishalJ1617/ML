```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
import warnings
warnings.filterwarnings('ignore')
```

```python
fp_df=pd.read_csv('Cosmetics.csv',index_col=0)
fp_df
```

```python
fp_df=fp_df.astype(bool)
```

```python
fp_df
```

##  Using Gradio for interface

```python
def gen_rules(min_sup,min_coef):
    #create frequent items
    itmesets=apriori(fp_df,min_support=min_sup,use_colnames=True)
    
    #And convert into rules
    rules=association_rules(itmesets,metric='confidence',min_threshold=min_coef)
    rules=rules[['antecedents','consequents','support','confidence','lift']]
    rules=rules[rules['lift']>1]
    return rules.sort_values(by='lift',ascending=False)
```

```python
import gradio as gr
```

```python
demo=gr.Interface(gen_rules,
                 inputs=[gr.Slider(value=0.01,step=0.01, minimum=0.0001,maximum=1),
                        gr.Slider(value=0.01,step=0.01, minimum=0.0001,maximum=1)],
                 outputs='dataframe')
```

```python
if __name__ == "__main__":
    demo.launch()
```
