```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
arrest_df=pd.read_csv('USArrests.csv',index_col=0)
```

```python
arrest_df
```

```python
scaler=StandardScaler()
```

```python
scaled_arrest_data=pd.DataFrame(scaler.fit_transform(arrest_df),columns=arrest_df.columns)
scaled_arrest_data
```

```python
eps_range=[0.2,0.4,0.6,1]
min_sample=[2,3,4,5]
result=[]
```

```python
for i in eps_range:
    for j in min_sample:
        dbscan=DBSCAN(eps=i,min_samples=j)
        dbscan.fit(scaled_arrest_data)
        distinct=len(np.unique(dbscan.labels_))
        if(distinct>2):
            scaled_arrest_data['cluster']=dbscan.labels_
            new_data=scaled_arrest_data[scaled_arrest_data.cluster!=-1]
            score=silhouette_score(new_data.iloc[:,:4],new_data.iloc[:,4])
            result.append([i,j,distinct,score])
```

```python
result
```
