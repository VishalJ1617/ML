```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
protein_df=pd.read_csv('Protein.csv',index_col=0)
```

```python
protein_df
```

```python
scaler=StandardScaler()
```

```python
scaled_protein_data=scaler.fit_transform(protein_df)
```

```python
km=KMeans(n_clusters=3, random_state=23)
km.fit(scaled_protein_data)
scores_arrest.append(km.inertia_)
```

```python
prcomp=PCA()
```

```python
scores=prcomp.fit_transform(scaled_protein_data)
```

```python
prcomp.explained_variance_
```

```python
df_PCs=pd.DataFrame({'PC1':scores[:,0],'PC2':scores[:,1],'cluster':km.labels_})
```

```python
sns.scatterplot(data=df_PCs,x='PC1',y='PC2',hue='cluster',palette='bright')
plt.show()
```
