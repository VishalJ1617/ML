```python
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
milk=pd.read_csv('milk.csv',index_col=0)
```

```python
scaler=StandardScaler()
milk_scaled=scaler.fit_transform(milk)
milk_scaled=pd.DataFrame(milk_scaled,columns=milk.columns,index=milk.index)
milk_scaled
```

```python
clust_DB=DBSCAN(eps=1,min_samples=2)
clust_DB.fit(milk_scaled)
clust_DB.labels_
```

```python
milk_scaled['Clust']=clust_DB.labels_
milk_scl_inliers=milk_scaled[milk_scaled['Clust']!=-1]
milk_scl_inliers
```

```python
silhouette_score(milk_scl_inliers.iloc[:,:-1],milk_scl_inliers.iloc[:,-1])
```

```python
eps_range=[0.2,0.4,0.6,1]
mp_range=[2,3,4,5]
score=[]
```

```python
for i in eps_range:
    for j in mp_range:
        clust_DB=DBSCAN(eps=i,min_samples=j)
        clust_DB.fit(milk_scaled.iloc[:,:5])
        distincts=len(np.unique(clust_DB.labels_))
        if distincts>2:
            milk_scaled['Clust']=clust_DB.labels_
            milk_scl_inliers=milk_scaled[milk_scaled['Clust']!=-1]
            a=silhouette_score(milk_scl_inliers.iloc[:,:-1],milk_scl_inliers.iloc[:,-1])
            score.append([i,j,distincts,a])             
score 
```

```python
df_results=pd.DataFrame(score,columns=['eps','mp','n_clust','sil'])
df_results.sort_values(by="sil",ascending=False,inplace=True)
df_results.iloc[0]
```

```python
clust_DB=DBSCAN(eps=0.4,min_samples=2)
clust_DB.fit(milk_scaled)
```

```python
prcomp=PCA()
scores=prcomp.fit_transform(milk_scaled)
```

```python
df_PCs=pd.DataFrame({'PC1':scores[:,0],'PC2':scores[:,1],'cluster':clust_DB.labels_})
df_PCs
```

```python
sns.scatterplot(data=df_PCs,x='PC1',y='PC2',hue='cluster',palette='bright')
plt.show()
```
