```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
fp_df=pd.read_csv('Faceplate.csv',index_col=0)
fp_df
```

```python
fp_df=fp_df.astype(bool)
```

```python
fp_df
```

```python
#create frequent items
itmesets=apriori(fp_df,min_support=0.2,use_colnames=True)
```

```python
#And convert into rules
rules=association_rules(itmesets,metric='confidence',min_threshold=0.6)
rules=rules[['antecedents','consequents','support','confidence','lift']]
```

```python
rules=rules[rules['lift']>1]
```

```python
rules.sort_values(by='lift',ascending=False)
```

##  Using gradio for user interface

```python
def gen_rule(min_sup,min_thres):
    #create frequent items
    itmesets=apriori(fp_df,min_support=min_sup,use_colnames=True)
    #And convert into rules
    rules=association_rules(itmesets,metric='confidence',min_threshold=min_thres)
    rules=rules[['antecedents','consequents','support','confidence','lift']]
    rules=rules[rules['lift']>1]
    return rules.sort_values(by='lift',ascending=False)
```

```python
import gradio as gr
```

```python
demo=gr.Interface(gen_rule,inputs=[gr.Slider(minimum= 0.0001,maximum=1,value=0.1,step=0.1),
                                  gr.Slider(minimum= 0.0001,maximum=1,value=0.1,step=0.1)],
                 outputs='dataframe')
```

```python
if __name__=='__main__':
    demo.launch()
```
