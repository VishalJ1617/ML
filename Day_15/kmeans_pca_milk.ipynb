```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
milk_df=pd.read_csv('milk.csv',index_col=0)
milk_df
```

```python
scaler=StandardScaler()
```

```python
scaled_milk=scaler.fit_transform(milk_df)
```

```python
km=KMeans(n_clusters=3, random_state=23)
km.fit(scaled_milk)
```

```python
prcomp=PCA()
scores=prcomp.fit_transform(scaled_milk)
```

```python
prcomp.explained_variance_
```

```python
df_PCs=pd.DataFrame({'PC1':scores[:,0],'PC2':scores[:,1],'cluster':km.labels_})
df_PCs
```

```python
sns.scatterplot(data=df_PCs,x='PC1',y='PC2',hue='cluster',palette='bright')
plt.show()
```

```python
plt.figure(figsize=(12,6))
for i in np.arange(0,milk_df.shape[0]):
    plt.text(scores[i,0],scores[i,1],list(milk_df.index)[i])
    
sns.scatterplot(data=df_PCs,x='PC1',y='PC2',hue='cluster',palette='bright')
plt.show()
```
