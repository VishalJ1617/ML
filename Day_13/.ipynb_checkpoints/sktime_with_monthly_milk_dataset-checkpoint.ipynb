## Time series with sktime

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from sktime.utils.plotting import plot_series
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.forecasting.ets import AutoETS
from sktime.forecasting.base import ForecastingHorizon
from sktime.performance_metrics.forecasting import mean_squared_error
from sktime.forecasting.exp_smoothing import ExponentialSmoothing
import warnings
warnings.filterwarnings("ignore")
```

 sktime have all the methods that have been taught in the time series with sklearn module

```python
milk_df=pd.read_csv('monthly-milk-production-pounds-p.csv',index_col=0)
```

```python
milk_df.head()
```

```python
idx=pd.to_datetime(milk_df.index).to_period("M")
milk_df.index=idx
```

```python
plot_series(milk_df);
```

```python
y=milk_df['Milk']
y_train,y_test=temporal_train_test_split(y,test_size=12)
print(y_train.shape, y_test.shape)
```

```python
plot_series(y_train,y_test,labels=['y_train','y_test']);
```

```python
fh=ForecastingHorizon(y_test.index,is_relative=False)
```

## Exponential smoothing

1. intializing the exponential smoothing
2. fitting model on the training set
3. predicting from the model

```python
forecaster=ExponentialSmoothing(trend="add",seasonal='additive',sp=12)
forecaster.fit(y_train)
y_pred=forecaster.predict(fh)
plot_series(y_train,y_test,y_pred,labels=['y_train','y_test','y_pred'])
plt.show()
```

```python
mean_squared_error(y_test,y_pred,square_root=True)
```

# Auto Ets

```python
forecaster=AutoETS(auto=True,sp=12)
forecaster.fit(y_train)
print(forecaster.summary())
```

```python
y_pred=forecaster.predict(fh)
plot_series(y_train,y_test,y_pred,labels=['y_train','y_test','y_pred'])
plt.show()
```

```python
mean_squared_error(y_test,y_pred,square_root=True)
```
